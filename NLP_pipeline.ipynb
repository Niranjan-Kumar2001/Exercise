```python
import spacy
import textacy
dictionary = spacy.load('en_core_web_lg')
f = open('husky.txt','rt')
data = f.read()
doc = dictionary(data)
len(doc)
```

# 1.Sentence Segmentation 

```python
#breaking the doc into each sentences 
for sent in doc.sents:
  print(sent.text)
```

# 2.Tokenization

```python
#breaking  all the sentences into seperate words 
for token in doc:
  print(token.text)
```

# 3.Parts of speech tagging

```python
#Finding the figure of speech whether its a noun or verb or adjective
for token in doc:
  print(token.text, token.pos_, spacy.explain(token.pos_))
```

# 4.Lemmatization

```python
#printing the most basic form of the word
for token in doc:
  print(token.text, token.lemma_)
```

# 5.Stopword

```python
#printing all the stopwords 
spacy_stopwords = spacy.lang.en.stop_words.STOP_WORDS

for token in doc:
  if  token.is_stop:
    print(token.text)
```

# 6.Dependency parsing

```python
#printing out the relationship between each word to the other of a sentence
for token in doc:
  print (token.text, token.tag_, token.head.text, token.dep_)
```

```python
from spacy import displacy
displacy.render(doc, style='dep', jupyter=True, options={'distance':100})
```

# 7.Noun Phrases

```python
#printing out the noun phrases
for token in doc.noun_chunks:
  print(token)
```

# 8.Named Entity Recognition 

```python
#identifying what kind of word it is such as people or ORG or location 
for ent in doc.ents:
    print(ent.text, ent.label_, spacy.explain(ent.label_))
```

# 9.Coreference Resolution

```python
#figuring out all the words that are referring to the same entity.
stats = textacy.extract.semistructured_statements(doc,entity="Husky", cue='be') 
print('This is all we know about Husky')
for state in stats:
  subject, verb, fact = state 
  print(f"-  {fact}")
```
